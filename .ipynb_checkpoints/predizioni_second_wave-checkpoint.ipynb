## Estrapolazione second wave

```python
# Y  la porzione di time-serie da stimare
Y=trend['totale_positivi']['Italia'][sw_start:]
```

## Fitting con esponenziale


### Definizione funzione base

$y=y0+ae^{bt}$

```python
def basic_expo(t,y0,a,b):
    return y0+a*np.exp(b*t)    
```

### Definizione dello spoazio di ricerca

```python
dominio=np.array([[0,0,-1],
                 [500,2,2]])
```

## Fitting

```python
days_training=len(Y)

def error_measure(predicted,actual):
    return np.mean(np.abs(predicted-actual))+np.std(np.abs(predicted-actual))
    #return 0.5*np.mean(np.abs(predicted-actual))+0.5*np.mean(np.abs(predicted[-14:]-actual[-14:]))
    #return 0.7*np.mean(np.abs(predicted[:-30]-actual[:-30]))+0.15*np.mean(np.abs(predicted[-14:-7]-actual[-14:-7]))+0.15*np.mean(np.abs(predicted[-7:]-actual[-7:]))
    #return np.percentile(np.abs(predicted-actual),80)+np.mean(np.abs(predicted-actual))
    #return np.mean(np.abs(predicted-actual))
    
def errore_relativo(predicted,actual):
    er=[np.abs(predicted[i]-actual[i])/actual[i] for i in range(len(actual))]
    return np.nanmean(er)

np.random.seed(42)
    
fcf=FagaCurveFit()

fcf.fit(basic_expo, 
        np.arange(days_training),
        Y,
        dominio,
        ga_generations=1000, ga_population=200,print_info=100,
        error_function=error_measure,
        num_mutations=1)
```

```python
solution=fcf.opt_params
solution
```

### Simulazione

```python
simulation_days=np.arange(len(Y))
```

```python
y_pred=np.array(list(map(lambda x:basic_expo(x,*solution),simulation_days)))
```

```python
fig,ax=plt.subplots(1,figsize=(15,6))
ax.plot(simulation_days,y_pred,label='Fitted')
ax.plot(simulation_days,Y,label='Actual')
ax.grid()
ax.set_ylabel('Nuovi positivi')
ax.legend()
```
