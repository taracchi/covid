# Analisi relazione tamponi vs positivi COVID-19

Si analizzano i trend, sia a livello **regionale** che **naizonale**
- del numero di tamponi effettuati
- dei positivi
- della relazione fra i due fattori

Fonte dati: *Presidenza del Consiglio dei Ministri - Dipartimento della Protezione Civile*

https://github.com/pcm-dpc/COVID-19

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

from aidam.math_utils import moving_function_rev1

import warnings
warnings.filterwarnings('ignore')
```

## Parametri

```python
data_file='dpc-covid19-ita-regioni.csv'
```

## Lettura dati e preprocessing

```python
data=pd.read_csv(data_file)
print(data.columns)

giorni=pd.unique(data['data'])
day_counter=list(range(len(giorni)))
print('Giorni osservati: %d'%len(giorni))
regioni=pd.unique(data['denominazione_regione'])
```

## Analisi

### Evoluzione tamponi e infetti per regione

Per ciascuna regioneviene plottato nel tempo il numero **totale** di tamponi effettuati e di casi di infezione.

```python
regioni_reshaped=np.reshape(regioni,(7,3))
fig,ax=plt.subplots(7,3,sharey=False,figsize=(15,15))
fig.tight_layout(pad=2.0)


for r in range(7):
    for c in range(3):
        regione=regioni_reshaped[r,c]
        regional_data=data[data['denominazione_regione']==regione]
        ax[r,c].plot(day_counter,regional_data['tamponi'],'b',label='Tamp')
        ax[r,c].plot(day_counter,regional_data['totale_casi'],'r',label='TotInf')
        ax[r,c].set_title(regione)
        ax[r,c].legend()
```

### Evoluzione rate infetti per tampone

Plot a livello sia globale che regionale della percentualedi nuovi nuovi casi *positivi* **giornalieri** a fronte del numero di *tamponi* **giornalieri** effettuati.

#### Globale

```python
# calcolo il rate globale
rate_new_tamp=100*np.diff(data.groupby(['data'])['totale_casi'].sum())/np.diff(data.groupby(['data'])['tamponi'].sum())

# plot dato grezzo
plt.plot(day_counter[1:],
         rate_new_tamp,
         'bo-')
plt.ylabel('% Tamponi positivi')

plt.plot(day_counter[1:],
         moving_function_rev1(rate_new_tamp,np.mean,3,3),
         'r--')

plt.grid(which='major')
```

#### Regionale

```python
fig,ax=plt.subplots(7,3,sharey=False,figsize=(15,15))
fig.tight_layout(pad=2.0)

for r in range(7):
    for c in range(3):
        regione=regioni_reshaped[r,c]
        regional_data=data[data['denominazione_regione']==regione]
        ax[r,c].plot(day_counter[1:],
                     100*np.diff(regional_data['totale_casi'])/np.diff(regional_data['tamponi']),
                     'bo-')
        ax[r,c].set_ylabel('% Tamponi pos.')
        ax[r,c].grid()
        ax[r,c].set_title(regione)
```
