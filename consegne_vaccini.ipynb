## Parametri

```python
data_file='https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/consegne-vaccini-latest.csv'
```

## Lettura dati e preprocessing

```python
consegne_data=pd.read_csv(data_file)
print(consegne_data.columns)

# riordino in base alla data (che creo e faccio indice)
consegne_data['data'] = pd.to_datetime(consegne_data['data_consegna'])
consegne_data.index=consegne_data['data']
consegne_data=consegne_data.sort_index()

# creo una colonna col numero di giorni dalla prima consegna
zero_day=consegne_data['data'][0]
day_counter=[(dd-zero_day).days for dd in consegne_data['data']]
consegne_data['day_counter']=day_counter

giorni=pd.unique(consegne_data['data_consegna'])

print('Giorni osservati: %d'%len(giorni))
print('Da: %s'%giorni[0][0:10])
print(' A: %s'%giorni[-1][0:10])
regioni=pd.unique(consegne_data['area'])

#data[data['denominazione_regione']=='Lombardia'].head(5)
consegne_data[consegne_data['area']=='TOS']#.tail(15)
```

```python
consegne_ita=consegne_data.groupby(['data_consegna']).sum()
#consegne_ita[consegne_ita['fornitore']=='AstraZeneca']
print(consegne_ita)
#consegne_ita.iloc[data=='2021-03-01']
```

## Creazione struttura dati

```python
# indicatori osservati
indicatori=['fornitore', 'numero_dosi', 'data_consegna','day_counter']

# qui immetto tutti i trend per ciascuna regione
consegne={}
for regione in regioni:
    consegne[regione]={}
    for ind in indicatori:    
        consegne[regione][ind]=np.array(consegne_data[consegne_data['area']==regione][ind])
      
```
