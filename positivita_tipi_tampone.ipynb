# Analisi discovery tamponi

```python
import pandas as pd
import plotly.graph_objects as go
```

```python
data_file='https://raw.githubusercontent.com/pcm-dpc/COVID-19/master/dati-andamento-nazionale/dpc-covid19-ita-andamento-nazionale.csv'

data=pd.read_csv(data_file)
print(data.columns)
data.tail(10)
```

## Calcolo colonne accessorie (rate positività)

```python
data['perc_molecolare'] = 100*data['totale_positivi_test_molecolare'].diff() / \
    data['tamponi_test_molecolare'].diff()

data['perc_antigenico'] = 100*data['totale_positivi_test_antigenico_rapido'].diff() / \
    data['tamponi_test_antigenico_rapido'].diff()
```

```python
data['totale_positivi_test_molecolare'].diff()
```

```python
last_days=90

fig = go.Figure()

fig.add_trace(go.Scatter(x=data['data'][-last_days:],
                         y=data['perc_molecolare'][-last_days:],
                         name='Molecolare',
                         mode='lines+markers'))

fig.add_trace(go.Scatter(x=data['data'][-last_days:],
                         y=data['perc_antigenico'][-last_days:],
                         name='Antigenico',
                         mode='lines+markers'))


fig.update_layout(
    autosize=False,
    showlegend=True,
    title='Rate di positività per tamponi molacolari e antigenici',
    width=900,
    height=600)
fig.show()
```

```python

fig = go.Figure()

fig.add_trace(go.Scatter(x=data['data'][-last_days:],
                         y=data['perc_molecolare'][-last_days:] /
                         data['perc_antigenico'][-last_days:],
                         name='Molecolare',
                         mode='lines+markers'))


fig.update_layout(
    autosize=False,
    showlegend=False,
    title='Rapporto fra rate di positività per tamponi molacolari e antigenici',
    width=900,
    height=600)
fig.show()
```
