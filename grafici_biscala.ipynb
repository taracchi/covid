# Grafici a confronto

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import matplotlib.pyplot as plt

from aidam.math_utils import moving_function_rev1
#from aidam.miscellanea_utils import find_matlabstyle
#from aidam.aicharts import get_subplot_cells_sequence, auto_grid_shape
#import pickle as pkl
import numpy as np
from covid_lib import read_covid_data, find_day_index
```

```python
trend,regioni,giorni,giorno_settimana,popolazione,day_counter=read_covid_data('stored_data.pkl')
```

```python
trend.keys()
```

```python
trend1='nuovi_positivi'
trend2='perc_tamponi_positivi'

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=giorni, y=moving_function_rev1(trend[trend1]['Italia'],np.mean,3,3), name=trend1),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=giorni, y=moving_function_rev1(trend[trend2]['Italia'],np.mean,3,3), name=trend2),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="%s and %s"%(trend1,trend2)
)

# Set x-axis title
#fig.update_xaxes(title_text="Giorni")

# Set y-axes titles
fig.update_yaxes(title_text=trend1, secondary_y=False)
fig.update_yaxes(title_text=trend2, secondary_y=True)

fig.show()
```

## Stesso trend ma regioni diverse

```python
trend['nuovi_positivi']
```

```python
trend_name='nuovi_positivi'
regione1='Italia'
regione2='Umbria'

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=giorni, y=moving_function_rev1(trend[trend_name][regione1],np.mean,3,3), name=regione1),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(x=giorni, y=moving_function_rev1(trend[trend_name][regione2],np.mean,3,3), name=regione2),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="%s for %s and %s"%(trend_name,regione1,regione2),
    width=1000,height=500
)

# Set x-axis title
#fig.update_xaxes(title_text="Giorni")

# Set y-axes titles
fig.update_yaxes(title_text=regione1, secondary_y=False)
fig.update_yaxes(title_text=regione2, secondary_y=True)

fig.show()
```
