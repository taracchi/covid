```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generalized_sigmoid(t,a,b,M,alpha):
    y=M/(1+np.exp(-a*t+b))**alpha
    return y
```

```python
t=np.arange(0,100,4)
```

```python
a=0.03
b=1
M=50
alpha=1

test_pars=[a,b,M,alpha]
y_a=np.array(list(map(lambda x:generalized_sigmoid(x,*test_pars),t-10)))

test_pars=[0.01,b,M,alpha]
y_b=np.array(list(map(lambda x:generalized_sigmoid(x,*test_pars),t-10)))


plt.plot(t,y_a,'b')
plt.plot(t,y_b,'r--')
plt.show()
```

```python
a=1
b=1
M=50
alpha=30

test_pars=[a,b,M,alpha]
y_a=np.array(list(map(lambda x:generalized_sigmoid(x,*test_pars),t-10)))

plt.plot(t,y_a)
```
